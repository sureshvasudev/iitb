## Computation and Logical check

You are given a base salary and a bonus amount. Update the salary by adding the bonus using an assignment operator. Then, check if the updated salary is more than 50000 and the bonus is more than 5000.



```python
# Taking input
salary = int(input())
bonus = int(input())

# Write your code here
salary += bonus

eligible = (salary > 50000 ) and (bonus > 5000)

# Print the output
print(eligible)
```

## Printing String

Write a program that takes two variables representing a person's first name and last name, and prints the full name in the format "Last, First".

Sample Input:
John
Doe

Sample Output:
Doe, John


```python
# Taking input
first = input()
last = input()

# Write your code here
full_name = last + ","+first

# Print the output
print(full_name)
```

## Simple Interest

Given the principal amount, rate of interest, and time in years, calculate and print the simple interest using the formula SI = (P * R * T) / 100.

```python
# Taking input
P = float(input())
R = float(input())
T = float(input())

# Write your code here
SI = (P*R*T)/100

# Print the output
print(SI)
```

## Convert to Seconds
Write a program that takes hours, minutes, and seconds and prints the total time in seconds.

```python
# Taking input
hours = int(input())
minutes = int(input())
seconds = int(input())

# Write your code here
total_seconds = (hours * 3600) + (minutes * 60) + seconds
# Print the output
print(total_seconds)
```
